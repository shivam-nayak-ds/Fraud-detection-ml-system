## Removing Features with High Missing Values

Columns with 70% or more missing values were removed from the dataset.

```python
###  3 . Correlated features handle
df.select_dtypes(include='number').corr().head(20)
```

```python
corr = df.select_dtypes(include='number').corr().abs()

for col in corr.columns:
    for row in corr.index:
        if col != row and corr.loc[row,col] > 0.7:
            print(row, col)
```

Highly correlated features contain very similar information. When two features are strongly correlated, they can introduce redundancy in the model.

To handle this, we:

Calculate the correlation matrix for numerical features.

Identify feature pairs with correlation greater than a selected threshold (e.g., 0.7).

Remove one feature from each highly correlated pair.

This step helps:

 1. Reduce multicollinearity

 2.  Simplify the model

3.   Improve model generalization

After removing unpredictive and highly correlated features, the dataset becomes cleaner and more suitable for model training.

```python
## step 4-  fill Missing values
```

```python
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

### Filling Missing Values

The dataset contains missing values in both numerical and categorical features.
To prepare the data for machine learning models, these missing values need to be handled.

* **Numerical Features:** Missing values are filled using the **median** of the column.
* **Categorical Features:** Missing values are replaced with a new category called **"Unknown"**.

This ensures that the dataset has no missing values and is ready for further preprocessing and model training.
