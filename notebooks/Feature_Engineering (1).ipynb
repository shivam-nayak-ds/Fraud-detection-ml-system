```python
import numpy as np

X['TransactionAmt_Log'] = np.log1p(X['TransactionAmt'])
```

```python
X['TransactionAmt_decimal'] = X['TransactionAmt'] - X['TransactionAmt'].astype(int)
```

```python
X.head()
```

```python
# Time features
X['Transaction_hour'] = X['TransactionDT']% 86400 // 3600
X['Transaction_day'] = X['TransactionDT'] // 86400
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

# Numerical columns
num_cols = X.select_dtypes(include=['int64', 'float64']).columns


scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])


```

```python
X
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print('Before SMOTE' ,y.value_counts())

print('After SMOTE' , y_resampled.value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_resampled , y_resampled)

importance = pd.Series(rf.feature_importances_, index=X.columns)
print("Top 10 Features:")
print(importance.nlargest(10))

```

```python
top_50 = importance.nlargest(50).index
X_selected = X_resampled[top_50]
y_selected = y_resampled

print( X_resampled.shape)
print( X_selected.shape)
```
